### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
"""
 Se comienza por estimar las ventas de cada una de las tiendas.
"""

def ingresos_tiendas (tiendas):
  ingresos = tiendas["Precio"]
  ingresos_totales=sum(ingresos)
  return ingresos_totales

print(f"La tienda 1 tuvo un ingreso de: {ingresos_tiendas(tienda)}")
print(f"La tienda 2 tuvo un ingreso de: {ingresos_tiendas(tienda2)}")
print(f"La tienda 3 tuvo un ingreso de: {ingresos_tiendas(tienda3)}")
print(f"La tienda 4 tuvo un ingreso de: {ingresos_tiendas(tienda4)}")
```

# 2. Ventas por categoría

```python

from matplotlib import pyplot as plt

def ventas_por_categoria(tiendas,numero):



  electronicos=[]
  muebles=[]
  juguetes=[]
  electrodomesticos=[]
  articulos_hogar=[]
  instrumentos_musicales=[]
  deportes_diversion=[]
  libros=[]

  for i in range(len(tiendas)):
    if tiendas["Categoría del Producto"][i]=="Muebles":
      muebles.append(tiendas["Producto"][i])
    elif tiendas["Categoría del Producto"][i]=="Electrónicos":
      electronicos.append(tiendas["Producto"][i])
    elif tienda["Categoría del Producto"][i]=="Juguetes":
      juguetes.append(tiendas["Producto"][i])
    elif tiendas["Categoría del Producto"][i]=="Electrodomésticos":
      electrodomesticos.append(tiendas["Producto"][i])
    elif tiendas["Categoría del Producto"][i]=="Artículos para el hogar":
      articulos_hogar.append(tiendas["Producto"][i])
    elif tiendas["Categoría del Producto"][i]=="Instrumentos musicales":
      instrumentos_musicales.append(tiendas["Producto"][i])
    elif tiendas["Categoría del Producto"][i]=="Deportes y diversión":
      deportes_diversion.append(tiendas["Producto"][i])
    else:
      libros.append(tiendas["Producto"][i])

  print(f"""\n
  A continuación se observa la cantidad de productos vendidos por categoria de la tienda {numero}:

            *Electrónicos vendidos: {len(electronicos)}
            *Muebles vendidos: {len(muebles)}
            *Juguetes vendidos: {len(juguetes)}
            *Electrodomésticos vendidos: {len(electrodomesticos)}
            *Artículos para el hogar vendidos: {len(articulos_hogar)}
            *Instrumentos musicales vendidos: {len(instrumentos_musicales)}
            *Artículos relacionados con deportes y diversión: {len(deportes_diversion)}
            *Libros vendidos son: {len(libros)}

  Las categorías que más productos venden, se muestran en la siguiente gráfica: \n""")

  #Listas para graficar

  categorias = ["Electrónicos","Muebles","Juguetes","Electrodomésticos","Artículos para el hogar","Instrumentos musicales",
                "Deportes y diversión","Libros"]
  cantidades = [len(electronicos),len(muebles),len(juguetes),len(electrodomesticos),len(articulos_hogar),len(instrumentos_musicales),
                len(deportes_diversion),len(libros)]


  #Crear pares y ordenar pares para mostrar grafica de forma ordenada y descendente sin modificar valores originales
  pares_ordenados = sorted(zip(categorias, cantidades), key=lambda x: x[1], reverse=True)
  categorias_ordenadas, cantidades_ordenadas = zip(*pares_ordenados)

  #Dibujar gráfica
  plt.bar(categorias_ordenadas,cantidades_ordenadas, color=["b","b","b","k","k","k","k","k"])
  plt.xticks(rotation=90, ha='right')
  plt.title(f"Cantidad de productos vendidos por categoría.  Tienda {numero}\n",color ="r")
  plt.show()

ventas_por_categoria(tienda,1)
ventas_por_categoria(tienda2,2)
ventas_por_categoria(tienda3,3)
ventas_por_categoria(tienda4,4)
```

# 3. Calificación promedio de la tienda


```python
def calificacion_promedio(tiendas,numero):
  calificaciones = tiendas["Calificación"]
  promedio = sum(calificaciones)/len(calificaciones)
  print(f"La calificación de los clientes en promedio de la tienda {numero} es de: {round(promedio,2)}")

calificacion_promedio(tienda,1)
calificacion_promedio(tienda2,2)
calificacion_promedio(tienda3,3)
calificacion_promedio(tienda4,4)

#pendiente el decidir si mostrar o no las graficas
```

# 4. Productos más y menos vendidos

```python

def cantidad_de_productos_vendidos(tiendas,numero):
  productos = tiendas["Producto"].value_counts()
  productos_mas_vendidos = productos.head(10)
  productos_menos_vendidos =productos.tail(10)

  titulo=f"Tienda {numero}"
  print("\n",titulo.center(100,"-"),"\n")

  print(f"En este apartado se puede ver la cantidad de productos vendidos por la tienda {numero}, estos son:\n\n{productos}")

  print(f"\nLos productos más y menos vendidos se visualizan en las siguientes gráficas:\n")

  #grafica de productos mas vendidos
  plt.barh(productos_mas_vendidos.index,productos_mas_vendidos.values)
  plt.title(f"Productos más vendidos por la tienda {numero}\n",color ="r")
  plt.show()

  #grafica de productos menos vendidos
  plt.barh(productos_menos_vendidos.index,productos_menos_vendidos.values,color="y")
  plt.title(f"Productos menos vendidos por la tienda {numero}\n",color ="r")
  plt.show()

cantidad_de_productos_vendidos(tienda,1)
cantidad_de_productos_vendidos(tienda2,2)
cantidad_de_productos_vendidos(tienda3,3)
cantidad_de_productos_vendidos(tienda4,4)
```

# 5. Envío promedio por tienda

```python
def costo_de_envio(tiendas,numero):
  envio = tiendas["Costo de envío"]
  promedio_de_envio=envio.mean()
  print(f"El costo promedio de envío de la tienda {numero} es de: {round(promedio_de_envio,2)}")

costo_de_envio(tienda,1)
costo_de_envio(tienda2,2)
costo_de_envio(tienda3,3)
costo_de_envio(tienda4,4)
```
